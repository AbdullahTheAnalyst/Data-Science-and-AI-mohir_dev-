```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sonlar = np.arange(10)
sonlar
```

```python
# line graph chiqarish
plt.plot(sonlar)
plt.show()
```

```python
# x ning qiymatlari : 0.1 tashab 0 dan 4*pi gacha sonlar
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
# plt.figure(figsize = (15, 5)) -> umumiy chizma o'lchami
# plt.plot(x, y, linewidth = 2, linestyle = "--", color = "r", marker = "^") -> qalinligi, chiziq-uslubi, rangi, chiziq-turi
# plt.grid(linestyle = "--") -> orqa-fon uslubi
# plt.legend(["sin(x)", "y"], loc = 0, fontsize = "x-large") -> qisqacha ma'lumot berguvchi qism (0, 1, 2, 3)
plt.figure(figsize = (15, 5))
plt.plot(x, y, linewidth = 2, linestyle = "--", color = "r", marker = "^")
plt.title("sin(x) funksiyasining Line Graphda ko'rinishi:")
plt.xlabel("x o'qining qiymatlari")
plt.ylabel("y o'qining qiymatlari")
plt.grid(linestyle = "--")
plt.legend(["sin(x)", "y"], loc = 0, fontsize = "x-large")
plt.show()
```
