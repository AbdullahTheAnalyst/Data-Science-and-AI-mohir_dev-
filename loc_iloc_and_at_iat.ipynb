```python
import pandas as pd
import numpy as np
```

```python
uz = {2024:37.5, 2023:36.4, 2022:34.7, 2021:33.5, 2020:32.6}
kz = {2024:34.5, 2023:37.4, 2022:31.7, 2021:34.5, 2020:38.6}
tj = {2024:31.5, 2023:30.4, 2022:30.7, 2021:31.5, 2020:37.6}
kg = {2024:38.5, 2023:31.4, 2022:33.7, 2021:37.5, 2020:30.6}
tr = {2024:37.5, 2023:32.4, 2022:39.7, 2021:38.5, 2020:34.6}
df = pd.DataFrame ({
    "Uzbekistan" : uz,
    "Kazakistan" : uz,
    "Tajikistan" : uz,
    "Kirgizistan" : uz,
    "Turkmanistan" : uz
})
df
```

```python
# df uchun 2021 indexning qiymatlarini oliyapmiz. loc-> Location so'zidan kelgan
df.loc [[2021]]
```

```python
df.loc [[2021, 2020]]
```

```python
# bir vaqtning o'zida ham Row ham Column ajratib olsak bo'ladi.
df.loc [[2021, 2020], ["Uzbekistan", "Tajikistan"]]
```

```python
# iloc -> index bo'yicha topadi
# bir vaqtning o'zida ham Row ham Column ajratib olsak bo'ladi.
df.iloc [[0, 3], [1, 2]]
```

```python
df
```

```python
# at -> orqali biz cellni (row va column kesishmasi) ni elementini olishimiz mumkin
df.at [2022, "Kirgizistan"]
```

```python
# bular orqali arifmetik amallar ham bajarsak bo'ladi
# 2020-yil Tojikiston aholisi 2022-yildagi O'zbekiston aholisidan 2.1 mln ta inson kam ekan
df.at[2022, "Uzbekistan"] - df.at[2020, "Tajikistan"]
```

```python
# Index orqali cell ni elementini olish uchun iat-> dan foydalanamiz
# 2022-yil Qizg'iziston kesishmasi
df.iat [2, 3]
```
