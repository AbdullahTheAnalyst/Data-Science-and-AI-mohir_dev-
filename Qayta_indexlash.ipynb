```python
import pandas as pd
```

```python
data = ({
    "Yil" : [2024, 2023, 2022, 2021, 2020, 2019, 2018],
    "Aholi" : [33, 32, 30, 29, 28, 27, 29],
    "Temp" : [1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8],
    "Zichlik" : [22, 25, 24, 28, 29, 30, 20]
})
df = pd.DataFrame (data)
df
```

```python
# indexlarni qo'l bilan o'zgartirdik
df.index = [24, 23, 22, 21, 20, 19, 18]
df
```

```python
# Reindex - bu orqali biz teskari tartibda indexlarni o'rnini almashtirishimiz mumkin
ind = [24, 23, 22, 21, 20, 19, 18]
ind.sort()
ind
```

```python
# bu qayta tartiblaganimiz DF o'zgaruvchining o'ziga ta'sir qilmaydi.
# Buning uchun esa qayta saqlashimiz kerak.
df.reindex(ind)
```

```python
df = df.reindex (ind)
df
```

```python
# biz kiritgan ustunlar qoladi, qolgani esa o'chib ketadi
df3 = df.reindex (columns = ["Aholi", "Temp", "Zichlik"])
df3
```

```python
# tasodifiy 1 ta qiymat chiqarib beradi
df3.sample()
```

```python
# tasodifiy 3 ta qiymat chiqarib beradi
df3.sample(3)
```
