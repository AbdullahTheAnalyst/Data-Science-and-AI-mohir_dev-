```python
import pandas as pd
```

```python
data = ({
    "Yil" : [2024, 2023, 2022, 2021, 2020, 2019, 2018],
    "Aholi" : [33, 32, 30, 29, 28, 27, 29],
    "Temp" : [1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8],
    "Zichlik" : [22, 25, 24, 28, 29, 30, 20]
})
df = pd.DataFrame (data)
df
```

```python
# drop() funksiyasi orqali biz, index qatorni olib tashashimiz mumkin
# drop(5) -> 5 nomli index qatorni o'chirib tashadik
df.drop (5)
```

```python
# drop() funksiyasi orqali biz, index qatorni olib tashashimiz mumkin
# bu asosiy df ga ta'sir qilmaydi
# drop([5, 1]) -> 5 va 1 nomli index qatorni o'chirib tashadik
df.drop ([5, 1])
```

```python
# drop("variable", axis=1) funksiyasi orqali biz, Column qatorni olib tashashimiz mumkin.
df.drop ("Yil", axis = 1)
```

```python
# drop("variable", axis=1) funksiyasi orqali biz,
# Columndan Yil va Temp variableni o'chrib tashadik qatorni olib tashashimiz mumkin
df.drop (["Yil", "Temp"], axis = 1)
```

```python
# lekin bu o'zgarishlar asosiy df ta'sir qilmaydi, agar yangi o'zgaruvchi bilan saqlamasak
df
```

```python
# lekin o'z vaqtida save qilmoqchi bo'lsak, 1-usul yangi o'zgaruvchi yaratib saqlash
# 2-usul esa inplace = True qo'shishdir , lekin bundan foydalanish tafsiya qilinmaydi
df.drop ("Yil", axis = 1, inplace = True)
df
```
