```python
data = {
    'age':[10,20,30,40,50,60],
    'year':[2010,2000,1990,1980,1970,1960]
    }

obj = pd.DataFrame(data)
obj
```

```python
# Korrelyatsiya degani ikki data'ning bir-biriga bog'liqligi.
# -1 va 1 oralig'ida bo'ladi.
# -1 va 1 ya yaqin bo'lsa bog'liqlik shuncha yaxshi degani
# manfiy (negativ) chiqsa, bittasi o'sib, bittasi kamayishini bilishimiz mumkin
# musbat (activ) chiqsa, bittasi o'sib, ikkinchi ham o'sishini bilamiz
obj["age"].corr(obj["year"])
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df
```

```python
# yil bilan bosib o'tgan yo'lning nisbatini topamiz (moshina qancha yili eski bo'lsa, shuncha ko'p masofa bosib o'tgan)
# korrelyatsiya orqali predictive tasklarga juda muhim
df["year"].corr(df["mileage"])
```

```python
df["price"].corr(df["mileage"])
```
