**.dropna() – NaN qiymatlarni tashlab yuborish**

Mavjud bo'lmagan qiymatlarga ishlov berishning yagona, universal formulasi yo'q.

Bunday holatda sizning birinchi savolingiz "Nima uchun mu qiymat mavjud emas? Ma'lumot yozib olinmaganmi yoki umuman mavjud emasmi?" bo'lishi kerak. Misol uchun, siz so'rovnoma yordamida odamlar haqida ma'lumotlar jamlayapsiz, so'rovnomangizda "nikohdan o'tgan yilingiz" degan bo'lim bor. Tabiiyki, turmush qurmaghanlar uchun bu qiymat mabjud bo'lmaydi va bu xato emas. Aksincha, "tug'ilgan yilingiz" qiymati mavjud bo'lmasligi esa, ma'lumot mavjud lekin yozib olinmaganini anglatadi.

Data Science muhandisi sifatida siz mavjud bo'lmagan ma'lumotlarni topishga, buning imkoni bo'lmasa ularni approksimasiya qilishga (tahminiy qiymat topishga) harakat qilishingiz, buni ham iloji bo'lmasa bunday ustun (yoki qatorlardan) voz kechishni hal qilishingiz kerak bo'ladi.

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv",
                 index_col = 0)
df.head()
```

```python
df.shape
```

```python
# jadvaldagi NaN qiymatlar bo'lsa True, bo'lmasa False qiymat beradi
df.isnull()
```

```python
# jadvaldagi NaN qiymatlar bo'lsa False, bo'lmasa True qiymat beradi
df.notnull()
```

```python
# NaN qiymat bo'lgan elementlarning sonini beradi
df.isnull().sum()
```

```python
# NaN qiymat bo'lmagan elementlarning sonini beradi
df.notnull().sum()
```

```python
# rowslar soni
len(df)
```

```python
# umumiy rowsning necha foizi null qiymat ekanligini aniqlash uchun:
df.isnull().sum() / len(df) * 100
```

```python
# NaN qiymatlarni tushurib qoldirish uchun, lekin bu amalni kam null qiymat bo'lsa bajarish kerak
df.dropna()
```

```python
# columndan ham o'chirish mumkin faqat
df.dropna(axis = 1)
```

```python
# df ichidagi faqat biz hoxlagan NaN qiymati bor ustunning qiymatlarini o'chirib tashash
df.dropna(subset = ["price"])
```

```python
df.dropna(subset = ["horsepower"])
```

```python
df.dropna(subset = ["company", "engine-type"])
```

```python
# tepadagi ish orqali asosiy df ga ta'sir o'tkazmayapti, save qilish uchun alohida o'zgaruvchi yaratib unga kiritib qo'yishimiz zarur
# yoki inplace = True -> orqali amalga oshirishimiz mumkin
df.dropna(subset = ["company", "price"], inplace = True)
```

**.fillna() – NaN qiymatlarni toʻldirish**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv",
                 index_col = 0)
df.head()
```

```python
# NaN qiymatlilarni to'ldirib qo'yish afzal hisoblanadi
# bu kod orqali barcha null qiymatlar uchun 0 berildi, albatta bunaqa qo'yish tavsiya etilmyaydim lekin arifmetik
# amallar bajarish uchun yaxshi hisoblanadi
df.fillna(0)
```

```python
# creative bo'lib integerli qatorlarga mean olib yozib qo'ysak bo'ladi
np.mean(df["horsepower"])
```

```python
# creative bo'lib integerli qatorlarga mean olib yozib qo'ysak bo'ladi
# bu lug'at shaklida yoziladi
# bu kod orqali biz horsepower va length larning null qiymatlariga shu columnlarning o'rtacha qiymatlarini berdik
df.fillna({"horsepower" : np.mean(df["horsepower"]),
           "length" : np.mean(df["length"])
           })
```

```python
df.fillna({"company" : "yo'q"

})
```

```python
df["company"]
```

```python
# demak company ustunida null qiymatlari yo'q
"yo'q" in df["company"]
```

```python
df
```

```python
# null qiymatning o'zidan oldingi qiymat bilan to'ldirib ketish funksiyasi
# ffill -> forward fill
df.fillna(method = "ffill")
```

```python
# null qiymatning o'zidan keyingi qiymat bilan to'ldirib ketish funksiyasi
# bfill -> back fill
df.fillna(method = "bfill")
```

```python
df[df.company == "toyota"]
```

```python
df[df.company == "toyota"].fillna(method = "ffill")
```

```python
# umumiy hammma jadvaldagi nulllarni o'rtacha qiymatlarini yozish
df.fillna(df.mean())
```
