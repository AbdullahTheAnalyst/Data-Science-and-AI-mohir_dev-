**DISTRIBUTION - Taqsimot**
Taqsimot grafik - o'zgaruvchi (ustun)da qanday qiymatlar mavjudligi va ularning ehtimolliligini ko'rsatuvchi grafik

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/merc.csv")
df = df[df["year"] >= 2010]
df = df[df["transmission"] != "Others"]
df = df[df["fuelType"] != "Others"]
df.head()
```

```python
df.shape
```

```python
# price ustun qanday taqsimotda ekanligini analiz qilish
plt.figure(figsize = (15,5))
sns.histplot(data = df, x = "price", hue = "transmission")
plt.show() # eng ko'p cars $20000 atrofida ekan
```

```python
# .histplot bilan bir xil
sns.displot(data = df, x = "price", hue = "transmission", height = 6, aspect = 2)
plt.show() # eng ko'p cars $20000 atrofida ekan
```

```python
# har bir ustun bo'yicha tahlil qilmoqchi bo'lsang!
# .displot ustun tomoni column.larning har birini alohida tahlil qilsa bo'ladi
sns.displot(data = df, x = "price", col = "transmission", height = 6, aspect = 2)
plt.show()
```

```python
# taqsimotning zichligini line qilib beradi
sns.displot(data = df, x = "price", kind = "kde", height = 5, aspect = 2)
plt.show()
```

```python
# taqsimotning zichligini line qilib beradi 2-usul ham plot ham kde chiqaradi
sns.displot(data = df, x = "price", kde = True, height = 5, aspect = 2)
plt.show()
```

```python
# .displot umumiy bo'lganligi sababli, .kdeplot() ishlatishimiz ham mumkin
plt.figure(figsize = (15,5))
sns.kdeplot(data = df, x = "price", shade = True) # shade.ni faqat .kdeplot() ishlataolamiz
plt.show()
```

```python
plt.figure(figsize = (15,5))
sns.kdeplot(data = df, x = "price",  hue = "fuelType", shade = True) # shade.ni faqat .kdeplot() ishlataolamiz
plt.show()
```

```python
# masalan: $25000 narxidagi moshinalar soni 60% ini tashkil qiladi.
sns.displot(data = df, x = "price", kind = "ecdf", height = 5, aspect = 2)
plt.xticks(rotation = 90)
plt.grid(which = "both")
plt.minorticks_on()
plt.show()
```

```python
# 3-o'lchamli kde grafika
plt.figure(figsize = (15,5))
sns.kdeplot(data = df, x = "price",  y = "year", hue = "fuelType", shade = True) # shade.ni faqat .kdeplot() ishlataolamiz
plt.show()
```

**subplot yordamida 1 ta chizmada bir nechta grafiklarni chizishimiz mumkin.**

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
plt.subplot(2,1,1) # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 1-grafik
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.title('sin(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish

plt.subplot(2,1,2) # # 2 - jami qatorlar soni, 1 - jami ustunlar soni, 2-grafik
# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')
plt.title('cos(x) funksiya grafigi')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish

plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

subplots funksiyasi 2 ta obyekt qaytaradi:

Umumiy chizma obyekti
Chizma tarkibidagi grafiklar ro'yxati
Biz quyida, qulaylik uchun, bu ikkalasini fig (figure) va ax (axes) o'zgaruvchilariga saqlab olamiz:

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
# Umumiy chizma nomini beramiz:
fig.suptitle("Grafik 1")
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x=df['year'], y=df['price'], hue="fuelType", s=50)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,10))
# Umumiy chizma nomini beramiz:
fig.suptitle("Grafik 2")
sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1], data=df, x='mileage')
sns.scatterplot(ax=ax[1, 0], data=df, x=df['year'], y=df['price'], hue="fuelType", s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x=df['year'], y=df['mileage'], hue="fuelType", s=50)
# Har bir grafik uchun nom:
ax[0,0].set_title("Price distribution")
ax[0,1].set_title("Mileage distribution")
ax[1,0].set_title("Price vs Year correlation")
ax[1,1].set_title("Mileage vs Year correlation")
plt.show()
```
