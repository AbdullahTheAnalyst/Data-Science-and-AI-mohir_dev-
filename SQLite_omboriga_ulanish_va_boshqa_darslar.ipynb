```python
# database ga ulanishga bog'lama beriladi
db = sql.connect("chinook.db")
db
```

```python
# databasega ayrim ishlov berish (yozish, o'qish) uchun cursor o'zgaruvchi yaratib oliandi
cursor = db.cursor()
```

```python
# ma'lumotlar omborida jadvallar mavjud bo'ladi. Shularni korish uchun ajratib olishimiz kerak.
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table'")
# cursor db ichidagi barcha jadvallarni chiqaradi .fetchall orqali
cursor.fetchall()
```

```python
# ma'lumotlar ombori ichida ma'lumotlar jadval ko'rinishda saqlanadi
# bu parametr orqali jadvalning rowslar sonini olamiz
cursor.execute("SELECT COUNT (*) FROM artists")
cursor.fetchone()
```

```python
# cursor db ichidagi ayrim jadvalni chiqaradi .fetchone orqali
cursor.fetchone()
```

```python
# ma'lumotlar ichidagi jadvalning Columns ning nomlarini aniqlash uchun:
cursor.execute("PRAGMA table_info(customers)")
cursor.fetchall()
# integer -> sonli, Nvarchar -> tring ko'rinishidagi ma'lumot
# Pragma -> faqat Sqlite da mavjud
```

```python
cursor.execute("PRAGMA table_info(playlists)")
cursor.fetchall()
```

```python
cursor.execute("PRAGMA table_info(albums)")
cursor.fetchall()
```

```python
cursor.execute("SELECT COUNT (*) FROM 'albums'")
cursor.fetchall()
```

```python
# biz bunaqa ichidagi datalarni chiqarsak bizga juda ham noqulay bo'ladi.
# buni bartaraf qilishimiz uchun esa bizga jadval ko'rishida olishimiz kerak
cursor.execute("SELECT * FROM playlists")
cursor.fetchall()
```

```python
# ya'ni pandas.read_sql -> dan foydalanamiz
# biz qayerdagi jadval kerak va qaysi databasedan olinganligi kerak
df = pd.read_sql("SELECT * FROM customers", db)
df.head()
```

```python
# indexni o'zimizga qulayini qo'yishimiz mumkin.
df = pd.read_sql("SELECT * FROM customers", db, index_col = "CustomerId")
df.head()
```

```python
# LIMIT orqali jadvalning tepadan nechta rows ni chiqarish bo'ladi
# bu df.head(20) bilan bir xil hisoblanadi
df = pd.read_sql("SELECT * FROM customers LIMIT 20", db)
df
```

```python
# sql ichidagi (*) barchasi anglatgani uchun bizga kerakli ustunlar kerak bo'lsa faqat nomini yozib qo'yamiz
df1 = pd.read_sql("SELECT CustomerID, FirstName, LastName, Phone FROM customers", db, index_col = "CustomerId")
df1.tail()
```

```python
# df1 = pd.read_sql("SELECT CustomerID, FirstName, LastName, Phone FROM customers", db, index_col = "CustomerId")
# yuqoridagi kodimiz juda uzun bo'lib ketdi
# shuning uchun sqlite ichidagi kodimizni alohida o'zgaruvchiga saqlab ishlashimiz mumkin
sql_uchun = "SELECT CustomerID, FirstName, LastName, Phone FROM customers"
df2 = pd.read_sql(sql_uchun, db)
df2.head()
```

```python
# ORDER BY orqali tartiblaydi
sql_uchun = "SELECT CustomerID, FirstName, LastName, Phone FROM customers ORDER BY FirstName"
df2 = pd.read_sql(sql_uchun, db)
df2
```

```python
# agar bizga faqat ma'lumot kerak bo'lsa
# ya'ni customersdan Countryning ichidan takrorlanmas elementlarni ber deyapmiz
sql_country = "SELECT DISTINCT Country FROM customers"
cursor.execute(sql_country)
cursor.fetchall()
```

```python
# buni biz dataframe orqali olmoqchi bo'lsak
df4 = pd.read_sql(sql_country, db)
df4
```

```python
# endi biz filtr o'rnatishimiz mumkin
# faqat country qismidan USA ga tegishlilarni filtrdan chiqardik
sql_query = "SELECT * FROM customers WHERE country = 'USA'"
df6 = pd.read_sql(sql_query, db)
df6
```

```python
# faqat country qismidan USA ga tegishlilarni filtrdan chiqardik
sql_query = "SELECT * FROM customers WHERE country = 'USA' OR country = 'Argentina'"
df6 = pd.read_sql(sql_query, db)
df6
```

```python
# faqat country qismidan USA ga tegishlilarni filtrdan chiqardik
# ismi bo'yicha tartiblandi
sql_query = "SELECT * FROM customers WHERE country = 'USA' AND country = 'Argentina' ORDER BY "
df6 = pd.read_sql(sql_query, db)
df6
```

```python
# counry USA va State CA orqali filtr qo'yildi
sql_query = "SELECT * FROM customers WHERE country = 'USA' AND State = 'CA' ORDER BY FirstName"
df6 = pd.read_sql(sql_query, db)
df6
```
