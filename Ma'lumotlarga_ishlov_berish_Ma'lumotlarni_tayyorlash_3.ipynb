**.groupby()orqali guruhlash**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv",
                 index_col = 0)
df.head()
```

```python
df.company.unique()
```

```python
df[df.company == "toyota"]
```

```python
df.company.value_counts()
```

```python
# .yuqoridagi kodni filterlashning .groupby orqali bajarish
df.groupby("company").max()
```

```python
df.groupby("company").min()
```

```python
# endi ustunlarning bizga keraklisini olib natijalarni analiz qilishimiz mumkin
df.groupby("company").price.mean()
```

```python
df.groupby("company").price.max()
```

```python
df.groupby("company").price.sum()
```

```python
# bitta emas ikki va undan ortiq guruhga ajratishimiz mumkin
df.groupby(["company", "engine-type"]).price.mean()
```

```python
df.groupby(["company", "engine-type"]).max()
```

```python
# bir guruh orqali bir nechta funksiyalarni qo'llash
df.groupby("company").agg([min, max])
```

```python
# faqat bitta ustun orqali chiqarish
df.groupby("company").price.agg([min, max, np.median])
```

```python
df.groupby("body-style").price.agg([len, min, max])
```

```python
df.groupby("body-style").agg([len, min, max])
```

**Jadvaldan tasodifiy qiymatlarni olish**

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/MOCK_USER_DATA.csv",
                  index_col = 0)
df3.head()
```

```python
len(df3)
```

```python
# biz ayrim vaqtda tartibli turgan narsalarni aralashtirib yuborishimiz kerak bo'ladi
# jadvalimiz 100 ta bo'lgani uchun 100 yozamiz
# permutation bu aralashtirib yuboradi
random_list = np.random.permutation(100)
random_list
```

```python
# yuqoridagi tasodifiy indexlar orqali joylashtirish kerak bo'lsa
df3.take(random_list)
```
