```python
import pandas as pd
import numpy as np
from pandas import Series
```

```python
# dataframe ma'lumotlar tuzulmasining Series dan farqi shundaki, dataframening ustun va qatorlari bo'ladi
# dataframe asosan katta excel, csv fayllardan biz ochib olamiz, datalarni analiz qilish uchun

data = ({
    "Yil:" : [2024, 2023, 2022, 2021, 2020, 2019],
    "Aholi soni:" : [35.6, 33.4, 32.10, 31.24, 30.57, 29.6],
    "Temp:" : [1.2, 1.6, 1.5, 1.7, 1.9, 1.8]
})

df = pd.DataFrame (data)
df
```

```python
df = pd.DataFrame (data, columns = ["Yil:", "Temp:", "Aholi soni:"])  # biz dataframening variables joyini almashtirib qo'yishimiz mumkin
df
```

```python
df = pd.DataFrame (data, columns = ["Yil:", "Temp:", "Aholi soni:"], index = [24, 23, 22, 21, 20, 19])
# biz dataframening indexlar nomini ham o'zgartirib qo'yishimiz mumkin
df
```

```python
df = pd.DataFrame (data, columns = ["Yil:", "Temp:", "Aholi soni:", "YaIM"], index = [24, 23, 22, 21, 20, 19])
# biz qo'chimcha YAIM qo'shdik va uning qiymatlari yo'qligi uchun NaN chiqardi
df
```

```python
aholi = df ["Yil:"] # har doim Series qismi kerak bo'lsa tepada from pandas import Series esingdan chiqmasin
aholi
```
