```python
import pandas as pd
import numpy as np
```

```python
data = ({
    "Yil" : [2024, 2023, 2022, 2021, 2020, 2019, 2018],
    "Aholi" : [33, 32, 30, 29, 28, 27, 29],
    "Temp" : [1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8],
    "Zichlik" : [22, 25, 24, 28, 29, 30, 20]
})
df = pd.DataFrame (data)
df
```

```python
# dataframe o'zi Har bir variables Serieslardan tashkil topgan ma'lumotlar qatoridir.
aholi = df ["Aholi"]
aholi
```

```python
# Seriesligini tekshirib oldik
type (aholi)
```

```python
# dataframe ni ayni bir variableni DataFrame qilib olishimiz uchun biz quyidagicha kod yozamiz
aholi = df [["Aholi"]]
aholi
```

```python
# DataFrame bo'lganini tekshirib oldik
type (aholi)
```

```python
# menga aholi variabledan yuqoridan 4ta qiymat kerak bo'lsa:
aholi [0:4]
```
