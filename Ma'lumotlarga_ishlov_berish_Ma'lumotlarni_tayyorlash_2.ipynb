Guruhlash uchun biz aholi sonini guruhlarga bo'lib chiqaylik:

*   1 mlngacha (0-1000)
*   1-10 mlngacha (1000-10000)

*   10-30 mlngacha (10000-30000)

*   30-50 mlngacha (30000-50000)
*   50-100 mlngacha (50000-100000)


*   100-300 mlngacha (100000-300000)


*   300 mln-1.5 mlrdngacha (300000-1500000)








```python
# shu oralig'da mavjud bo'lgan o'zgaruvchi yaratib olamiz
number_pop = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
```

```python
population = df.pop2021
population
```

```python
# manashu datani guruhlarga ajratish uchun
guruh = pd.cut(population, number_pop)
guruh
```

```python
guruh.value_counts()
```

```python
# o'zimizning df ga alohida data qo'shish uchun
df["guruh"] = guruh
df
```

```python
df5 = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv",
                  index_col = 0)
df5.head()
```

```python
df5.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
```

```python
range = pd.cut(df5.price, bins)
range
```

```python
range.value_counts()
```

```python
# biz umumiy ma'lumotlarni olishimiz uchun oraliq qiymatlarni nomlar olishimiz mumkin
names = ["arzon cars", "o'rta cars", "qimmat cars"]
range = pd.cut(df5.price, bins, labels = names)
range.value_counts()
```

```python
# hozir bu range qilishimizni qo'lda bajardik, endi buni qo'lda emas, avtomatik qilishimiz mumkin
range = pd.cut(df5.price, 6)
range
```

```python
range = pd.cut(df5.price, 6)
range.value_counts()
```

```python
range = pd.cut(df5.horsepower, 4)
range.value_counts()
```

```python
# tepacagi .cut() metodi orqali range qiymatlari oralig'i juda beo'xshov edi
# lekin .qcut() orqali bir- biriga yaqin qiymat chiqaradigan oraliqdagini bajaradi
range = pd.qcut(df5.price, 3)
range.value_counts()
```

```python
range = pd.qcut(df5.horsepower, 4)
range.value_counts()
```
