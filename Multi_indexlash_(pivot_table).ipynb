```python
# bu Multi indexlash bizga juda katta database ga ishlov berish uchun kerak bo'ladi
# Odatiy Series to'plam
data = pd.Series(np.random.randn(9), index = ["a",
                                              "a",
                                              "a",
                                              "b",
                                              "b",
                                              "c",
                                              "c",
                                              "d",
                                              "d",
                                              ])
data
```

```python
data.index
```

```python
# buni multi - indexga alishtiramiz
# yuqori masshtabli db uchun guruhlashga juda yordam beradi
data1 = pd.Series(np.random.randn(9), index = [["a", "a", "a", "b", "b", "c", "c", "d", "d"],
                                              [1, 2, 3, 1, 3, 1, 2, 2, 3]])
data1
```

```python
data1.index
```

```python
# faqat c indexiga tegishli datani ajratib olamiz
data1["c"]
```

```python
# ko'p indexini ajratib olish
data1["c" : "d"]
```

```python
data1["a" : "c"]
```

```python
# barcha 3-indexga tegishli elementlarni ajratib olish
data1.loc[ : , 3]
```

```python
# barcha b va c indexdagi elementlar
data1.loc[["b", "c"]]
```

```python
# unstack va stack usullari juda foydali hisoblanadi
# unstack -> bu multi-indexdan dataframega sochib yuboradi
# unstack -> sochilgan Seriesni df ko'rinishiga qaytaradi
data1.unstack()
```

```python
# stack -> df.ni yig'ilgan Series.ga o'giradi
data1.unstack().stack()
```

```python
df = pd.read_csv("/content/Practicum_Datasets/world_cities_air_water_quality.csv")
df.head()
```

```python
# har doim data cleaning.dan oldin albatta data haqida ma'lumotlari haqida info ol
df.info()
```

```python
# ishni column name.ni to'g'irlashdan boshlaymiz ("" ichida yozilgan xatosi)
index_col = ["City", "Region", "Country", "AirQuality", "WaterPollution"]
df.columns = index_col
df
```

```python
# Region va Country ichida hamma element "" ichida yozilgan va buni to'g'irlashimiz kerak
# .map(lambda x: x.replace('"', '')) -> " ning o'rnida bo'sh qoldirish uchun dastur
df["Region"] = df["Region"].map(lambda x: x.replace('"', ''))
df
```

```python
# Region va Country ichida hamma element "" ichida yozilgan va buni to'g'irlashimiz kerak
# .map(lambda x: x.replace('"', '')) -> " ning o'rnida bo'sh qoldirish uchun dastur
df["Country"] = df["Country"].map(lambda x: x.replace('"', ''))
df
```

```python
# yuqorida update qilingan data.ni yangi csv file.ga save qilish kerak
# index = False ->boshqa index paydo bo'lishini oldini oladi
df.to_csv("Updated_dataset", index = False)
```
