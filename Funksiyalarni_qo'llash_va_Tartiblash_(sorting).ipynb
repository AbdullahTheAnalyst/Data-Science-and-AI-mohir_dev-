```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/usa_cars.csv', index_col=0)
df.head()
```

```python
# dataframe uchun sorting
df.sort_index(axis = 1, inplace = True)
```

```python
# dataframe uchun teskari tartibdagi sorting
df.sort_index(ascending = False)
```

```python
# bitta qiymat bo'yicha esa by metodi kerak
df.sort_values(by = "price")
```

```python
# ikki va undan ortiq values bo'yicha sorting va bunga ham by metodi kerak
# birinchi price orqali keyin esa mileage values orqali sorting bo'ldi
df.sort_values(by = ["price", "mileage"])
```

```python
# ikki va undan ortiq values bo'yicha sorting va bunga ham by metodi kerak
# birinchi price orqali keyin esa mileage values orqali teskari sorting bo'ldi
df.sort_values(by = ["price", "mileage"], ascending = False)
```
