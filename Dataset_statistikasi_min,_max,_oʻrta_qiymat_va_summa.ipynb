```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/uzbekistan.csv', index_col=0)
df.columns = ['maydoni','aholisi']
df
```

```python
# dfning umumiy summasini topdik
df.sum()
```

```python
df2 = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/gm_sales.csv", index_col=0)
df2
```

```python
# variable bo'yicha summasini topdik
df2.sum(axis = "columns")
```

```python
# Rows bo'yicha summasini topdik
df2.sum(axis = "rows")
```

```python
# o'rta arifmetigi
df2.mean()
```

```python
# o'rta arifmetigi columnni
df2.mean(axis = "columns")
```

```python
# o'rta arifmetigi rowsni
df2.mean(axis = "rows")
```

```python
df.mean()
```
