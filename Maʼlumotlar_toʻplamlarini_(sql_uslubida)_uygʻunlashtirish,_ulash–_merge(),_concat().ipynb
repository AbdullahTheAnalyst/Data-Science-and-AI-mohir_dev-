```python
a = pd.DataFrame({
    "Transport" : ["poyezd", "avtomobil", "kema"],
    "Tezligi" : [70, 180, 45]
    })
a
```

```python
b = pd.DataFrame({
    "Transport" : ["poyezd", "avtomobil", "tramvay"],
    "Narxi" : [98000, 10000, 65000]
    })
b
```

```python
# merge funksiyasi -> bizga faqat ikki db.da mavjud bo'lgan qiymatlarini oladi
# bu inner-join hisoblanadi
merged_inner = pd.merge(a, b) # yoki
merged_inner = pd.merge(a, b, how = "inner")
merged_inner
```

```python
# bu outer-join hisoblanadi va barcha elementlar chiqadi
merged_outer = pd.merge(a, b, how = "outer")
merged_outer
```

```python
# left-join -> chapgi birlashish ya'ni umumiy elementlarning faqat chapdagisi olinadi
merged_left = pd.merge(a, b, how = "left")
merged_left
```

```python
# right-join -> o'nggi birlashish ya'ni umumiy elementlarning faqat o'ngdagisi olinadi
merged_right = pd.merge(a, b, how = "right")
merged_right
```

```python
# ikkin db.ning ustunlari nomi mos bo'lsa yuqoridagi kodimiz ishlaydi
# masalan
a = pd.DataFrame({
    "Transport_turi" : ["poyezd", "avtomobil", "kema"], # Transportni Transport_turi bo'lsin,
    "Tezligi" : [70, 180, 45]
    })
a
```

```python
pd.merge(a, b, how = "inner")
```

```python
# No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False
# ikki db.ning column nomlari o'xshamasa quyidagicha bajaramiz
merged = pd.merge(a, b, left_on = "Transport_turi", right_on = "Transport")
merged
```

```python
# bittasini o'chirish kerak bo'lsa:
merged = pd.merge(a, b, left_on = "Transport_turi", right_on = "Transport")
merged_drop = merged.drop(columns = ["Transport"])
merged_drop #yoki
del merged["Transport"]
merged
```

**Index ustida birlashtirish**

```python
a = a.set_index("Transport_turi")
b = b.set_index("Transport")
merged_index = pd.merge(a, b, left_index = True, right_index = True, how = "inner")
merged_index
```

```python
merged_index = pd.merge(a, b, left_index = True, right_index = True, how = "outer")
merged_index
```
