```python
# biz bo'sh database yaratib qo'yapmiz
db = sql.connect("newdb.db")
cursor = db.cursor()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/Electricity_Production_By_Source.csv")
df.head()
```

**SQL jadval yaratish, ma ºlumotlar qo ªshish**

```python
# SQL da jadval yaratish quyidagicha bo'ladi
# """ """ -> orqali yozilgan kodlar ketma-ket pastdan joylashadi
# jadval ustuni nomi bilan uning mohiyatini ham beriladi
# double o'nlik son (double = real)
# not null -> qiymat bo'lishi shart hisoblanadi
# primary key esa asosiy kalit hisoblanadi
# db.commit() ->(tastiqlash) o'zgarishlarni saqlab ma'lumotlar omboriga yoziladi
query = """
      CREATE TABLE IF NOT EXISTS worldpop(
        rank INT,
        country TEXT PRIMARY KEY,
        density DOUBLE,
        population DOUBLE NOT NULL,
        area INT NOT NULL
      );
"""
cursor.execute(query)
db.commit()
```

```python
# databasega joylandimi yo'qmi tekshirish uchun
cursor.execute("SELECT name FROM sqlite_master WHERE type = 'table'")
print(cursor.fetchall())
```

```python
cursor.execute("PRAGMA table_info(worldpop)")
cursor.fetchall()
```

**Ma'lumot qo'shish**

**SQL buyruqlarini f-string yordamida yozish**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population.csv", index_col = "rank")
df.head()
```

```python
query = """
      CREATE TABLE IF NOT EXISTS worldpop(
        rank INT,
        country TEXT PRIMARY KEY,
        density DOUBLE,
        population DOUBLE NOT NULL,
        area INT NOT NULL
      );
"""
cursor.execute(query)
db.commit()
```

```python
uz = df[df.country == "Uzbekistan"]
uz
```

```python
# f string yordamida yozishning qulaylik tomoni biz har gal qo'lda datani kiritib o'tirmaymiz
# faqatgina o'zgaruvchilarni o'zgartirishimiz kifoya qiladi
rank = uz.index.values[0]
country = uz.at[rank, "country"]
density = uz.at[rank, "density"]
population = uz.at[rank, "pop2021"]
area = uz.at[rank, "area"]
print(f"{rank}, '{country}', {density}, {population}, {area}")
```

```python
query = f"""
    INSERT INTO worldpop VALUES ({rank}, '{country}', {density}, {population}, {area})
"""
cursor.execute(query)
db.commit()
```

```python
cursor.execute("SELECT * FROM worldpop")
cursor.fetchall()
```
