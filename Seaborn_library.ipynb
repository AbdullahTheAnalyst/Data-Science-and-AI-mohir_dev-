```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# seaborn kutubxonasi matplotlib ustiga qurilgan visualization library hisoblanadi
# seaborn ancha takomillashgan library hisoblanadi
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/museum_visitors.csv", index_col = 0)
df.head()
```

```python
df.info()
```

```python
# chiziqli plot
sns.lineplot(data = df)
plt.show()
```

```python
# matplotlib.dan farqlash
plt.plot(df)
plt.show()
```

```python
plt.figure(figsize = (16, 6))
plt.title("Muzeylarga tashriflar soni:")
plt.xlabel("Sanasi:")
plt.ylabel("Odamlar soni:")
plt.xticks(rotation = 90) # 90 gradusga aylantirish
plt.grid()
sns.lineplot(data = df)
plt.show()
```

```python
# faqat bitta turdagi muzeyni chiqarish mumkin
plt.figure(figsize = (16, 6))
plt.title("America Tropical Interpretive Center muzeyiga tashriflar soni:")
plt.xlabel("Sanasi:")
plt.ylabel("Odamlar soni:")
plt.xticks(rotation = 90) # 90 gradusga aylantirish
plt.grid()
sns.lineplot(data = df["America Tropical Interpretive Center"])
plt.show()
```
